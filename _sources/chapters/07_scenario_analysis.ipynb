# Scenario Analysis and Planning Tools

Wastewater infrastructure planning requires long-term vision, data-driven projections, and structured decision-making frameworks to anticipate demographic shifts, climate impacts, regulatory changes, and economic constraints. This chapter presents analytical tools and scenario modeling approaches to support strategic planning for Vermont's wastewater systems in the context of the state's broader economic and demographic challenges.

## Introduction

Vermont faces a complex set of economic and demographic challenges that directly affect wastewater infrastructure planning. According to the [Vermont Competitiveness Dashboard](https://vtfuturesproject.org/vermont-competitiveness-dashboard/)—a comprehensive compilation of economic indicators developed by the Vermont Futures Project—Vermont ranks near the bottom nationally across multiple critical metrics: #49 in population change (losing 215 people in 2024), #46 in housing permits (2,654 units vs. 7,500+ needed), #51 in economic momentum, and #51 in effective business tax rates.

These trends create unique planning challenges for wastewater utilities:

- **Declining and aging population**: Vermont's population declined from 2023 to 2024 (one of only three states to lose population), with the state ranking #50 for dependency ratio and median age. Fewer working-age adults translate to shrinking ratepayer bases, declining flow projections, and challenges financing capital improvements.

- **Housing scarcity constraining growth**: Vermont needs to triple its housing production rate to meet the Economic Action Plan goal of 350,000 occupied homes by 2035 (currently 285,548). Inadequate wastewater capacity or unavailability of sewer connections can limit housing development and economic growth, while overbuilding capacity risks stranded assets if growth doesn't materialize.

- **High cost environment**: Vermont ranks #43 in cost of living, #42 in electricity prices, and #49 in public school spending per pupil, making infrastructure investments more expensive and competing with other community priorities for limited resources.

- **Business climate challenges**: Vermont's ranking of #38 in CNBC's Top States for Business (despite #1 for Quality of Life) reflects workforce, tax, and regulatory constraints that affect industrial/commercial wastewater customers and economic development potential.

- **Fiscal constraints**: The state's high tax burden (#42 in tax competitiveness, #51 in business tax rates) and restrictive fiscal policies (#49 in ALEC-Laffer Economic Outlook) limit available public funding for infrastructure grants and loans, placing greater burden on local utilities and ratepayers.

Effective wastewater planning in this environment requires:

- **Scenario-based projections**: Modeling multiple demographic, economic, and climate futures rather than assuming linear growth
- **Capacity right-sizing**: Balancing service reliability with financial sustainability to avoid overbuilding or underbuilding infrastructure
- **Climate adaptation**: Incorporating flood risk, precipitation intensity changes, and operational stress from extreme weather
- **Cost-benefit frameworks**: Rigorous economic analysis to prioritize investments and justify rate increases
- **Regional collaboration**: Exploring shared services, regionalization, and coordinated planning to achieve economies of scale
- **Decision support tools**: Structured methods for comparing alternatives and communicating trade-offs to stakeholders and regulators

This chapter presents analytical approaches to navigate these challenges and support strategic planning for resilient, affordable, and sustainable wastewater systems.

## Population and Demand Projections

Vermont's demographic trajectory is one of the most critical uncertainties for wastewater planning. The Vermont Futures Project's Economic Action Plan sets a long-term goal of 802,000 residents by 2035 (currently 648,493), requiring substantial reversal of current population decline. Wastewater utilities must plan for multiple scenarios that account for this uncertainty.

### Vermont's Demographic Challenge

Vermont's 2024 population change ranked #49 out of 50 states, with a net loss of 215 residents. Key demographic factors affecting wastewater demand include:

- **Net domestic migration**: While Vermont ranks #31 for net domestic and international migration (partially offset by international arrivals), the state struggles to attract and retain working-age adults
- **Aging population**: Vermont ranks #50 for dependency ratio and median age, with fewer residents in the prime 25-44 age cohort (#50 nationally); aging residents typically generate less wastewater per capita and create fiscal stress on utility rate structures
- **College-bound student retention**: Vermont ranks #50 in retention of college-bound students attending in-state schools, and #12 in import/export ratio, indicating significant youth out-migration despite quality higher education institutions
- **Educational attainment**: Vermont ranks #3 nationally for educational attainment of adults 25+, suggesting quality-of-life attractiveness that could support future in-migration if economic conditions improve

### Scenario Development

Wastewater planners should develop at least three demographic scenarios reflecting uncertainty in Vermont's economic future:

#### Low/Decline Scenario (Continuation of Current Trends)

Assumes Vermont's population continues slow decline or stagnation similar to 2020-2024 trends:
- **Population growth rate**: -0.1% to 0% annually
- **Wastewater demand**: Flat to declining in aggregate, with regional variation (urban/resort growth, rural decline)
- **Planning implications**: Excess treatment capacity, declining rate base, deferred capital investments, focus on asset management and efficiency rather than expansion
- **Risk**: Stranded assets from past overbuilding; difficulty financing even essential rehabilitation; declining service quality

#### Medium/Stable Scenario (Modest Recovery)

Assumes Vermont achieves modest population stabilization and selective growth through targeted economic development:
- **Population growth rate**: +0.3% to +0.5% annually (split between natural increase, domestic migration, international arrivals)
- **Wastewater demand**: Slow growth concentrated in job centers (Chittenden County, college towns, resort areas); stable or declining in rural areas
- **Planning implications**: Capacity expansions targeted to growth nodes; collection system extensions to accommodate infill development and housing; regionalization to serve growth areas efficiently
- **Risk**: Misallocation of capacity if growth doesn't occur as projected; under-investment if growth exceeds projections

#### High/Aspirational Scenario (Economic Action Plan Goals)

Assumes Vermont successfully implements Economic Action Plan strategies to achieve 802,000 residents by 2035:
- **Population growth rate**: +1.5% to +2.0% annually (requires aggressive in-migration of working-age adults and families)
- **Wastewater demand**: Substantial growth across multiple regions; new housing (targeting 350,000 occupied homes) drives collection system expansion and treatment capacity additions
- **Planning implications**: Major capital program to add capacity at WWTFs, expand collection systems, and upgrade aging infrastructure; CSO elimination accelerated; indirect discharge conversions to centralized treatment
- **Risk**: Overbuilding if growth targets not met; financing challenges for large capital programs; construction cost escalation and supply chain constraints

### Regional Variation

Statewide projections mask significant regional differences:

- **Chittenden County**: Likely continues to capture disproportionate share of growth regardless of statewide scenario; Burlington/Essex/Winooski WWTFs may need capacity expansions or process optimization; satellite communities (Hinesburg, Richmond, Shelburne) face extension/connection decisions
- **Resort/recreation areas**: Stowe, Killington, Stratton, and similar destinations may see continued second-home growth and tourism-driven demand independent of statewide population trends; seasonal flow variation complicates capacity planning
- **College towns**: Middlebury, Montpelier, Johnson, Castleton, and others benefit from stable institutional anchors but face student housing demand volatility
- **Rural towns**: Many small communities face depopulation, aging customer bases, and declining flows; consolidation/regionalization may be necessary to maintain financial viability
- **Border areas**: Communities near interstate borders (New Hampshire, Massachusetts, New York) may benefit from spillover growth if neighboring states offer economic advantages

### Wastewater Flow Projection Methods

Translate population scenarios to wastewater flows using:

- **Per capita flow factors**: Residential (50-70 gpcd), commercial/institutional (varies by type), industrial (site-specific); adjust for water conservation trends, appliance efficiency gains
- **Employment-based models**: For communities with significant commercial/industrial base, project flows based on job growth forecasts (Vermont Competitiveness Dashboard shows #45 ranking for establishment opening/closing rates)
- **Housing-unit method**: Flows = (occupied housing units) × (persons/household) × (gpcd) + commercial/industrial flows; ties directly to housing production targets (7,500+ units/year needed to meet 2035 goal)
- **Land-use planning**: Integrate municipal comprehensive plans, Act 250 permits, and designated growth areas to project spatial distribution of demand
- **I/I reduction**: Account for infiltration/inflow mitigation efforts that reduce wet-weather flows and increase available capacity without treatment expansion

### Non-Demographic Demand Drivers

Beyond population, consider:

- **Water conservation**: Statewide water efficiency initiatives, appliance standards, and leak reduction programs may reduce per capita flows 10-15% over 20 years, offsetting some population-driven demand growth
- **Economic development**: Major employers (e.g., GLOBALFOUNDRIES expansions, data centers, food processing) can add significant industrial load independent of population growth
- **Climate impacts**: More intense precipitation increases wet-weather I/I and CSO frequency; drought and heat waves concentrate flows in summer tourism season
- **Regulatory changes**: Tightening nutrient limits may require capacity reductions (to increase detention time for treatment) even if physical hydraulic capacity exists

### Implications for Planning Horizons

- **Short-term (5-10 years)**: Base capital plans on medium/stable scenario; identify early indicators (housing permits, employment trends, school enrollment) to adjust if trends shift toward low or high scenario
- **Mid-term (10-20 years)**: Maintain flexibility to scale investments up or down; modular/phased expansion designs allow adaptation as scenarios clarify
- **Long-term (20-30 years)**: Reserve land and planning approvals for potential expansions under high scenario; avoid irreversible decisions (e.g., building in floodplains, undersizing collection mains) that constrain future options
