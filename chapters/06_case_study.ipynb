### Treatment Types and Phosphorus Controls

Lake Champlain basin facilities employ various treatment approaches, with progressive tightening of phosphorus limits driving technology upgrades:

- **Chemical phosphorus precipitation**: Alum, ferrous chloride, or ferric chloride addition in primary or secondary treatment to precipitate phosphorus; widely implemented at Burlington, Montpelier, Rutland, Middlebury, and other major plants
- **Tertiary filtration**: Sand, cloth, or membrane filters polish effluent to achieve low TSS and phosphorus; common at facilities with stringent TMDL-driven limits
- **Biological phosphorus removal**: Limited use in Vermont due to cold climate challenges; some facilities operate enhanced biological P removal in warmer months
- **SBR technology**: Sequencing batch reactors at Stowe and other resort areas handle seasonal flow variation while maintaining nutrient removal
- **Lagoon and aerated pond systems**: Used in smaller communities (Hardwick, Richford) with seasonal phosphorus limits
- **Indirect discharge/land application**: Some resort and institutional facilities (ski areas, schools) use spray irrigation or rapid infiltration to reduce surface water discharge and leverage soil phosphorus attenuation

### Flow and Loading Profiles

Major facilities in the basin vary widely in design capacity and loading:

- **Large municipal WWTFs (>5 MGD)**: Burlington Main, Rutland, Montpelier, St. Johnsbury (Connecticut basin)—collectively treat the majority of basin flow and phosphorus load
- **Medium municipal WWTFs (1-5 MGD)**: Essex Junction, Winooski, Middlebury, Vergennes, Morrisville—significant contributors requiring TMDL compliance
- **Small municipal (<1 MGD)**: Numerous small towns and fire districts; individually small loads but cumulatively important
- **Industrial high-volume/low-strength**: GLOBALFOUNDRIES, IBM—high flow but process water with different pollutant profiles than domestic wastewater
- **Seasonal variation**: Resort area facilities (Stowe, Killington region) experience 2-4x flow variation from winter tourism

### Phosphorus Reduction Achievements and Ongoing Challenges

Since the Lake Champlain TMDL establishment, point-source phosphorus reductions have been substantial:

- **Burlington Main**: Reduced annual phosphorus loading by >50% through chemical precipitation, tertiary filtration, and CSO controls; one of the largest TMDL success stories
- **Rutland, Montpelier, Middlebury**: Major capital investments in phosphorus removal technologies, achieving sub-1.0 mg/L effluent concentrations in summer months
- **Basin-wide progress**: Collective wasteload allocations have driven facility upgrades, operational optimization, and enhanced monitoring, demonstrating measurable load reductions tracked through Vermont Clean Water Initiative reports

Ongoing challenges include:

- **Small-system capacity**: Limited financial and technical capacity to implement advanced treatment at small rural facilities
- **CSO/SSO nutrient pulses**: Combined systems still contribute phosphorus during wet weather despite substantial CSO control investments
- **Climate variability**: More intense storms and changing snowmelt patterns affect treatment efficiency and nutrient delivery timing
- **Aging infrastructure**: Collection system rehabilitation needed to reduce I/I and maximize treatment plant efficiency

## Phosphorus Management

Phosphorus is the primary nutrient of concern for Lake Champlain, driving algal blooms, cyanobacteria outbreaks, water quality impairment, and threats to recreation, drinking water, and aquatic habitat. Managing phosphorus loading from wastewater treatment facilities has been central to Lake Champlain restoration efforts for over three decades.

### Historical Context and Interstate Coordination

The Lake Champlain Basin Program (LCBP) has coordinated phosphorus reduction strategies across three jurisdictions since the early 1990s. The [LCBP's Phosphorus Reduction Strategy](https://www.lcbp.org/our-goals/clean-water/nutrients-and-cyanobacteria/phosphorus-reduction-strategy/) established a structured, science-based approach to phosphorus management through three critical steps:

#### 1. Establishment of Phosphorus Concentration Targets (1993)

In 1993, New York, Vermont, and Québec signed a Water Quality Agreement committing to a consistent approach to phosphorus management. The agreement:

- Defined in-lake phosphorus concentration criteria (standards) for Lake Champlain
- Established phosphorus concentration targets for each of the thirteen lake segments
- Was renewed in 2010 to maintain long-term commitment
- Provides the foundation for all subsequent phosphorus reduction efforts

These segment-specific targets recognize that Lake Champlain is not a uniform water body—shallow embayments like Missisquoi Bay and St. Albans Bay are more sensitive to nutrient enrichment than the deep Main Lake, requiring tailored management approaches.

#### 2. Development of Phosphorus Budget and Load Reduction Strategy (1994)

Vermont and New York completed a comprehensive diagnostic-feasibility study to:

- Measure point-source (wastewater treatment facilities) and nonpoint-source (agriculture, stormwater, developed lands) phosphorus loads
- Develop a whole-lake phosphorus budget quantifying sources and sinks
- Model how load reductions translate to in-lake concentration improvements
- Identify the magnitude of reductions needed to achieve concentration targets

Key findings from the phosphorus budget:

- **Current annual loads**: Approximately 859 metric tons of phosphorus enter Lake Champlain annually (as of 2018 State of the Lake report)
- **Target load**: 427 metric tons per year to achieve in-lake concentration goals
- **Required reduction**: ~432 metric tons annually—approximately 50% reduction from current loads
- **Nonpoint dominance**: Most excess phosphorus comes from nonpoint sources (agriculture, stormwater, streambank erosion)
- **Point-source progress**: Loads from wastewater treatment facilities are now well below established targets for this sector, reflecting decades of facility upgrades and operational improvements

This phosphorus budget approach—linking external loads to in-lake responses—provided the scientific foundation for setting achievable, measurable reduction targets.

#### 3. Cost-Optimization and Reduction Strategy Selection (1995-1996)

In 1995, Holmes and Artuso developed a cost-optimization procedure to identify the most cost-effective combinations of point- and nonpoint-source reductions for achieving in-lake phosphorus criteria. The optimization:

- Considered costs of reductions from agricultural lands, urban stormwater, and wastewater treatment
- Accounted for how phosphorus reductions in one lake segment affect other segments (via hydrodynamic transport)
- Allowed ranking of reduction scenarios by cost-effectiveness
- Informed a bistate negotiation process to allocate reductions fairly

In early 1996, representatives from Vermont, New York, and US EPA selected a load reduction strategy that:

- Targets 12 of 19 watersheds for phosphorus reductions (focusing on those with the greatest impact on impaired lake segments)
- Commits each state to reduce the difference between existing (1995) loads and target loads by at least 25% per five-year period for 20 years
- Allows states flexibility to choose the appropriate mix of point- and nonpoint-source actions within each watershed
- Permits load target adjustments as long as in-lake concentration goals are met and each state's allowable loads remain fixed
- Is contingent on availability of federal and/or state funding

This phased, 20-year reduction strategy recognized that achieving phosphorus targets would require sustained, multi-decade investment and adaptive management.

### Wastewater Treatment Facility Role

Point sources—primarily municipal and industrial wastewater treatment facilities—historically contributed a significant fraction of Lake Champlain's phosphorus load. Through aggressive facility upgrades, permit limit tightening, and operational optimization from the 1990s through present:

- **Load reductions achieved**: Point-source loads have been reduced to well below sector targets, representing one of the major success stories of Lake Champlain restoration
- **Technology upgrades**: Major facilities (Burlington, Rutland, Montpelier, Middlebury) installed chemical phosphorus precipitation (alum, ferric chloride), tertiary filtration, and enhanced monitoring
- **Permit stringency**: NPDES permits progressively reduced phosphorus concentration limits (many facilities now target <0.8 mg/L summer effluent phosphorus, with some achieving <0.5 mg/L)
- **CSO controls**: Combined sewer overflow programs in Burlington, Montpelier, Rutland, Middlebury, Vergennes, and other cities reduced wet-weather phosphorus pulses
- **Seasonal considerations**: Summer phosphorus limits (April-October) address the period of highest algal growth and lowest dilution flows

Despite this progress, wastewater facilities remain critical to Lake Champlain protection because:

- **Reliability**: Point sources provide consistent, measurable, year-round loads that can be controlled through permits
- **Local impacts**: Facilities discharging to shallow bays (Missisquoi, St. Albans, Burlington) have disproportionate influence on local water quality even when basin-wide loads are small
- **Growth accommodation**: Future population and economic growth will require wastewater infrastructure to handle increased loads without compromising phosphorus targets
- **Adaptive management**: As nonpoint-source reductions prove difficult and climate change alters hydrology, wastewater facilities may face additional reduction pressure

### Remaining Challenges

While wastewater point sources have largely met their reduction goals, overall Lake Champlain phosphorus reduction lags behind targets due to:

- **Nonpoint-source persistence**: Agricultural runoff, streambank erosion, and urban stormwater continue to deliver excess phosphorus; these diffuse sources are harder to control than point sources
- **Implementation gaps**: Funding and staffing constraints limit the pace of agricultural best management practice adoption, stormwater infrastructure upgrades, and river corridor restoration
- **Climate impacts**: More intense precipitation events increase runoff and erosion, counteracting some reduction efforts; warmer water temperatures exacerbate algal bloom risk even at lower phosphorus concentrations
- **Legacy phosphorus**: Decades of phosphorus accumulation in lake sediments and floodplain soils provide an internal source that can delay recovery even after external loads are reduced

### Interstate and International Coordination

Effective phosphorus management requires coordination across political boundaries:

- **Vermont-New York cooperation**: Joint TMDL development, shared monitoring networks, coordinated permit strategies, and mutual accountability for load reduction targets
- **Vermont-Québec Missisquoi Bay Agreement**: Addresses the highly impaired Missisquoi Bay/Baie Missisquoi, which is influenced by sources in both jurisdictions; separate agreement with specific bay-focused actions
- **Lake Champlain Basin Program**: Facilitates communication, provides technical and financial support, tracks progress through State of the Lake reporting, and maintains public engagement
- **Federal support**: US EPA oversight of NPDES permits, Section 319 nonpoint-source grants, and Clean Water State Revolving Fund support infrastructure investments

For detailed information on phosphorus reduction strategies, targets, and progress, see the [Lake Champlain Basin Program's Phosphorus Reduction Strategy](https://www.lcbp.org/our-goals/clean-water/nutrients-and-cyanobacteria/phosphorus-reduction-strategy/) and the [LCBP State of the Lake](http://sol.lcbp.org/) reporting.
