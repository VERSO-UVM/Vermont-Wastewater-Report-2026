## Flood Risk Assessment

Vermont's history of devastating floods—from the 1927 flood to Tropical Storm Irene in 2011 and the July 2023 flood—underscores the critical vulnerability of wastewater infrastructure to flooding. Understanding flood risk and implementing protective strategies are essential for maintaining public health, environmental protection, and system operability during and after extreme weather events.

### Vermont's Flood Context

Vermont's geography and climate make it particularly susceptible to flooding. Intense rainstorms, rapid snowmelt, ice jams, and remnants of tropical systems can overwhelm rivers and tributaries, inundating low-lying areas where many wastewater treatment facilities, pump stations, and collection systems are located. Historical flood events have repeatedly demonstrated that wastewater infrastructure—especially facilities constructed decades ago before modern flood hazard mapping—is at significant risk.

The 1927 flood remains Vermont's most devastating on record, destroying infrastructure across the state. Tropical Storm Irene (2011) caused $733 million in damage statewide, inundating wastewater treatment plants, washing out collection lines, and disrupting service to thousands of residents. More recently, intense rainfall events in July 2023 again flooded facilities across central and northern Vermont, requiring emergency repairs and highlighting ongoing vulnerability.

### Vermont Flood Ready Atlas

The [Vermont Flood Ready Atlas](https://floodready.vermont.gov/assessment/vt_floodready_atlas) provides a comprehensive online mapping tool for assessing flood risk to critical infrastructure, including wastewater facilities. The Atlas integrates:

- **FEMA Digital Flood Insurance Rate Maps (DFIRMs)**: Special Flood Hazard Areas (SFHAs) with 1% annual chance of flooding (100-year floodplain), floodways, and base flood elevations
- **River Corridor data**: Mapped for streams with watersheds ≥2 square miles, identifying areas needed for natural floodplain and channel migration
- **Watershed boundaries**: Uphill areas contributing runoff during storms
- **Critical facilities**: Locations of wastewater treatment plants, pump stations, and other essential infrastructure
- **National Flood Insurance Program (NFIP) participation**: Communities enrolled in flood insurance and floodplain management

Users can identify which wastewater facilities are located within SFHAs, floodways, or river corridors, assess cumulative risk across a watershed, and prioritize mitigation investments. The Atlas allows municipalities and operators to overlay facility locations with flood hazard layers, generate printable maps for planning, and track FEMA map updates as new flood studies are completed.

### Flood Risk Disclosures and Insurance

Vermont law requires property sellers to disclose flood zone status and flooding history to buyers, using FEMA Flood Insurance Rate Maps to determine whether a property lies within a designated Special Flood Hazard Area. For wastewater facilities and collection system assets in SFHAs, flood insurance is often mandatory if federally backed financing is involved, and strongly recommended regardless of financing source.

Understanding flood insurance requirements is critical for wastewater utilities:

- **NFIP policies**: Cover buildings and contents up to specified limits, with premiums based on flood zone, elevation, and building characteristics
- **Elevation certificates**: Required to establish accurate flood risk and premium rates, documenting the lowest floor elevation relative to base flood elevation
- **Coverage gaps**: Standard policies have limits and may not cover full replacement cost for large treatment facilities; excess coverage or catastrophe bonds may be necessary
- **Claims and mitigation**: Filing flood claims can increase future premiums, incentivizing proactive mitigation (elevating equipment, flood-proofing structures, relocating vulnerable components)

Property owners and municipalities should consult the [Vermont Flood Ready Assessment](https://floodready.vermont.gov/assessment) tools and work with regional floodplain managers to ensure adequate insurance coverage and understand disclosure obligations.

### Facility Vulnerability

Wastewater infrastructure elements at particular risk include:

- **Treatment plants in floodplains**: Many older WWTFs were sited near rivers for effluent discharge, placing them in SFHAs; floodwaters can inundate electrical systems, damage blowers/pumps, contaminate chemicals, and interrupt operations
- **Pump stations**: Low-lying pump stations serving collection systems can be submerged, causing sewage backups and SSOs if backup power or flood barriers are inadequate
- **Collection lines and manholes**: Floodwaters infiltrate sewers, overwhelming capacity and causing CSOs/SSOs; inflow surges can hydraulically overload treatment plants
- **Critical electrical and control systems**: Even brief flooding can destroy SCADA systems, motor control centers, and transformers, requiring costly replacement and extended downtime
- **Access and egress**: Flooded roads prevent staff access for emergency operations and chemical/fuel deliveries

FEMA flood maps identify the extent of 100-year and 500-year flood zones, but climate change and land-use changes can make historical flood frequency estimates unreliable; facilities should plan for events exceeding mapped flood elevations.

### Mitigation Strategies

Reducing flood vulnerability requires a combination of structural and non-structural measures:

- **Elevation and flood-proofing**: Raise electrical panels, blowers, generators, and chemical storage above base flood elevation; construct flood walls or berms around vulnerable facilities; install check valves to prevent backflow
- **Equipment redundancy and backup power**: Maintain portable pumps, backup generators with elevated fuel storage, and redundant critical systems to maintain operations during floods
- **Early warning and emergency response**: Integrate with National Weather Service flood forecasting, Vermont Flood Ready monitoring, and local emergency management to trigger pre-storm preparations (sandbagging, equipment shutdown, chemical containment)
- **Siting and design standards**: Apply Vermont's Flood Resilience Standards and River Corridor Protection to new construction and major upgrades, requiring facilities to be sited outside floodplains and river corridors where feasible, or designed to withstand flooding
- **Asset relocation**: Evaluate long-term viability of facilities in high-risk zones; for chronic flood exposure, relocating plants or major components may be more cost-effective than repeated repairs and emergency response

### Community Flood Resilience

Vermont's [Flood Ready initiative](https://floodready.vermont.gov/) promotes community-wide resilience by encouraging municipalities to:

- Adopt and enforce local floodplain bylaws exceeding NFIP minimums
- Participate in the Community Rating System (CRS) to reduce flood insurance premiums through enhanced floodplain management
- Protect river corridors and floodplains as natural flood storage and attenuation
- Develop Local Hazard Mitigation Plans identifying critical infrastructure vulnerabilities and prioritizing mitigation projects
- Qualify for Emergency Relief and Assistance Fund (ERAF) support by demonstrating floodplain management and river corridor protection

Wastewater utilities should coordinate with municipal planners, regional planning commissions, and state floodplain managers to integrate facility resilience into broader community flood preparedness and recovery planning.

### Data and Resources

Key resources for assessing wastewater facility flood risk:

- **[Vermont Flood Ready Atlas](https://floodready.vermont.gov/assessment/vt_floodready_atlas)** (bit.ly/floodatlas): Interactive mapping of flood hazards, river corridors, and critical infrastructure
- **[FEMA Map Service Center](https://msc.fema.gov/)**: Official FIRMs, FIRMettes, and flood study data
- **[Vermont Community Reports](https://floodready.vermont.gov/assessment/community_reports)**: Municipality-level summaries of flood hazard mapping status, NFIP participation, and ERAF eligibility
- **[Regional Floodplain Managers](http://dec.vermont.gov/watershed/floodplains)**: Technical assistance on flood hazard interpretation, permitting, and mitigation design
- **[Understanding Flood Risk Disclosures and Insurance in Vermont](https://landifyr.com/understanding-flood-risk-disclosures-and-insurance-in-vermont-a-comprehensive-guide/)**: Comprehensive guide to disclosure requirements, NFIP policies, and flood insurance claims

By leveraging these tools and adopting proactive flood risk reduction measures, Vermont's wastewater utilities can enhance resilience, protect water quality, and ensure continuity of service during increasingly frequent and severe flood events.
