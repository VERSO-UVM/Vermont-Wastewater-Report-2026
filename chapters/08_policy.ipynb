## Federal Regulatory Framework

### Clean Water Act (CWA)

The Clean Water Act is the primary federal law governing water pollution control in the United States.

Key provisions affecting wastewater systems:
- National Pollutant Discharge Elimination System (NPDES) permits
- Technology-based effluent limitations
- Water quality-based effluent limitations
- Total Maximum Daily Loads (TMDLs)

### EPA Regulations

Environmental Protection Agency regulations that apply to Vermont wastewater facilities:

- 40 CFR Part 122 - NPDES Permit Program
- 40 CFR Part 133 - Secondary Treatment Standards
- 40 CFR Part 503 - Biosolids Management

## Vermont State Regulations

### Vermont Water Quality Standards

Vermont has established water quality standards to protect designated uses of state waters.

Key standards include:
- Numeric criteria for conventional pollutants
- Narrative criteria for toxic substances
- Anti-degradation policy
- Special protection for Outstanding Resource Waters

### Vermont DEC Wastewater Program

The Vermont Department of Environmental Conservation (DEC) administers the state's wastewater program.
